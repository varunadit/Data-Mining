```python
######### Code for Task 2 #############

file = open('dataset.csv', 'r') 
list_values = file.readlines()
new_list=[]
for _ in list_values:
    new_list.append(_.strip())
file.close() 

final_list=[]
for _ in new_list:
    final_list.append(_.split(','))

te = TransactionEncoder()
te.fit(final_list)
te_ary=te.transform(final_list)

df=pd.DataFrame(te_ary, columns=te.columns_)

df
```

```python
print ('Case 1 : At least 30% support')
frequent_itemsets1 =fpgrowth(df, min_support=0.40, use_colnames=True)
frequent_itemsets1
```

```python
rules = association_rules(frequent_itemsets1, metric="confidence", min_threshold=0)
rules = rules.sort_values('confidence', ascending=True)
rules#.head()
```

#### The above rules are generated for purchases with support of atleast 40%, and minimum confidence 0%. 

In this case, the support threshold is set to 40%, a frequent itemset is generated for the items that occur in at least 40% of all transactions in the database.

However, upon building association rules, we see that there is no association between the two most frequent items of the transactions.

```python
print ('Case 2 : At least 10% support')
frequent_itemsets2=fpgrowth(df,min_support=0.1,use_colnames=True)
frequent_itemsets2
```

```python
rules = association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.5)
rules = rules.sort_values('confidence', ascending=True)
rules#.head()
```

#### The above rules are generated for purchases with support of atleast 10%, and minimum confidence 50%. 
In this case, the support threshold is set to 10%, a frequent itemset is generated for the items that occur in at least 40% of all transactions in the database.

Upon building association rules, we see that there is an association rule that gives us :
(sandwich loaves, vegetables) --> (poultry), confidence=0.5

This means that when a person does purchase sandwich loaves, and vegetables, then they will also purchase poultry with a confidence of 0.500 or 50% of the time.


```python
print ('Case 3 : At least 6% support')
frequent_itemsets3=fpgrowth(df,min_support=0.06,use_colnames=True)
frequent_itemsets3
```

```python
rules = association_rules(frequent_itemsets3, metric="confidence", min_threshold=0.3)
rules = rules.sort_values('confidence', ascending=False)
rules#.head()
```

#### The above rules are generated for purchases with support of atleast 6%, and minimum confidence 30%.

Upon building association rules, we see that there is an association rule that gives us :
(paper towels) --> (pork, vegetables), confidence=0.300242

This means that when a person does purchase paper towels, then they will also purchasepork, and vegetables with a confidence of 0.300242 or about 30.02% of the time.
